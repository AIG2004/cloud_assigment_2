```python
import pandas as pd
import numpy as np
df=pd.read_csv('books.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#dropping duplicates and nulls
df=df.dropna()
df=df.drop_duplicates()
```

```python
result = df[df['title'].str.startswith('Harry Potter')]
result.describe()
```

```python
average_rating=result['average_rating'].mean()
average_rating
```

```python
maximum_books_count=result['books_count'].max()
max_sold_book=result.loc[result['books_count'].idxmax(),'title']
print('max sold book is',max_sold_book,':',maximum_books_count)
```
